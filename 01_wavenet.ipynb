```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
from tensorflow.keras.layers import Input, Conv1D, Activation, Multiply, Bidirectional, LSTM, Dropout, Dense, GRU
from tensorflow.keras.models import Model

# Load dataset
df = pd.read_csv(r'D:\A_NAUSHAD\E\Dataset\Book7.csv', parse_dates=['Date'], index_col=['Date'])

# Plot correlation matrix
correlation_matrix = df.corr()
plt.figure(figsize=(15, 15))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Plot')
plt.show()

# Copy dataset
dataset = df.copy()

# Define the WaveNet layer
def wavenet_layer(inputs, dilation_rate, filters=64, kernel_size=2):  
    tanh_out = Conv1D(filters=filters, kernel_size=kernel_size, dilation_rate=dilation_rate, padding='causal')(inputs)
    tanh_out = Activation('tanh')(tanh_out)
    
    sigmoid_out = Conv1D(filters=filters, kernel_size=kernel_size, dilation_rate=dilation_rate, padding='causal')(inputs)
    sigmoid_out = Activation('sigmoid')(sigmoid_out)
    

    output = Multiply()([tanh_out, sigmoid_out])
    return output

# Loop through each feature in the dataset
for a in range(len(df.columns)):
    b = df.iloc[:, a:a+1]
    features = df.values  # Use b.values for single feature
    targets = b.values

    # Train-test-validation split
    X_train, X_test, y_train, y_test = train_test_split(features, targets, test_size=0.15, random_state=1, shuffle=False)
    X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.15, random_state=1, shuffle=False)

    print("features:", features.shape)
    print("targets:", targets.shape)
    print("X_train:", X_train.shape)
    print("X_test:", X_test.shape)
    print("X_val:", X_val.shape)
    print("y_train:", y_train.shape)
    print("y_test:", y_test.shape)
    print("y_val:", y_val.shape)
    
    # Scaling
    scaler = MinMaxScaler()
    scaler.fit(X_train)
    X_train = scaler.transform(X_train)
    X_val = scaler.transform(X_val)
    X_test = scaler.transform(X_test)

    win_length = 10
    batch_size = 10
    num_features = features.shape[1]

    # Creating Timeseries Generators
    train_generator = TimeseriesGenerator(X_train, y_train, length=win_length, sampling_rate=1, batch_size=batch_size)
    val_generator = TimeseriesGenerator(X_val, y_val, length=win_length, sampling_rate=1, batch_size=batch_size)
    test_generator = TimeseriesGenerator(X_test, y_test, length=win_length, sampling_rate=1, batch_size=batch_size)
    
    # Model creation
    inputs = Input(shape=(win_length, num_features))
    x = wavenet_layer(inputs, dilation_rate=2)
   # x = wavenet_layer(x, dilation_rate=4)
  #  x = wavenet_layer(x, dilation_rate=8)
  #  x = Bidirectional(LSTM(64, activation='tanh', return_sequences=True))(x)
    #x = Bidirectional(LSTM(32, activation='tanh', return_sequences=True))(x)
    x = Bidirectional(LSTM(75, activation='tanh', return_sequences=False))(x)
    x = Dropout(0.1)(x)
    outputs = Dense(1)(x)

    model_CNNLSTM = Model(inputs=inputs, outputs=outputs)

    # Compile the model
    model_CNNLSTM.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.01), loss=tf.losses.MeanSquaredLogarithmicError())

    # Plot the model
    tf.keras.utils.plot_model(model=model_CNNLSTM, show_shapes=True)
    
    # Callbacks for early stopping and learning rate reduction
    lr_monitor = tf.keras.callbacks.ReduceLROnPlateau(monitor="val_loss", patience=3, factor=0.5, cooldown=1)
    early_stopping = tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True)

    # Model training
    with tf.device('/GPU:0'):
        prepared_model_CNNLSTM = model_CNNLSTM.fit(train_generator, 
                                                   validation_data=val_generator,
                                                   epochs=200, 
                                                   shuffle=False,  
                                                   callbacks=[lr_monitor, early_stopping])
        
    # Plot training and validation loss
    plt.plot(prepared_model_CNNLSTM.history["loss"], label="loss")
    plt.plot(prepared_model_CNNLSTM.history["val_loss"], label="val_loss")
    plt.legend(loc="best")
    plt.xlabel("No. Of Epochs")
    plt.ylabel("MSLE Loss")
    plt.show()
    
    # Create a DataFrame for future forecasting (e.g., hourly timestamps from 2023 to 2025)
    future_dates = pd.date_range(start='2023-01-29', end='2025-12-31', freq='H')
    df_future = pd.DataFrame(index=future_dates)

    # Forecast future values using the trained model
    future_generator = TimeseriesGenerator(features, np.zeros(len(features)), length=win_length, sampling_rate=1, batch_size=1)
    predicted_values_future = model_CNNLSTM.predict(future_generator)

    # Ensure the length of forecasted values matches the length of the DataFrame index
    df_future['PM2.5'] = predicted_values_future[:len(df_future)]

    # Save the forecasted values
    f = df_future['PM2.5']
    cities = pd.DataFrame(f)
    cities.to_csv('D:/A_NAUSHAD/E/RESULTS/FORE/train_wavenet_' + df.iloc[:, a:a+1].columns[0] + '_S_fore.csv')

    # Plot the forecasted values
    plt.figure(figsize=(12, 6))
    plt.plot(df_future.index, df_future['PM2.5'], label='Forecasted PM2.5')
    plt.xlabel('Hour')
    plt.ylabel('PM2.5')
    plt.legend()
    plt.title('One-Year Forecast of Hourly PM2.5')
    plt.show()
    
    # Save training loss values
    cities = pd.DataFrame(prepared_model_CNNLSTM.history["loss"])
    cities.to_csv('D:/A_NAUSHAD/E/RESULTS/LOSS/train_wavenet_' + df.iloc[:, a:a+1].columns[0] + '_S_loss.csv')

    # Save validation loss values
    cities = pd.DataFrame(prepared_model_CNNLSTM.history["val_loss"])
    cities.to_csv('D:/A_NAUSHAD/E/RESULTS/LOSS/val_wavenet_' + df.iloc[:, a:a+1].columns[0] + '_S_loss.csv')

    # Evaluate model on training set
    model_CNNLSTM.evaluate(train_generator, verbose=1)
    predictions_train = model_CNNLSTM.predict(train_generator)

    # Save training predictions
    cities = pd.DataFrame(predictions_train)
    cities.to_csv('D:/A_NAUSHAD/E/RESULTS/PRED/train_wavenet_' + df.iloc[:, a:a+1].columns[0] + '_S_pred.csv')

    # Evaluate model on test set
    model_CNNLSTM.evaluate(test_generator, verbose=1)
    predictions_test = model_CNNLSTM.predict(test_generator)

    # Save test predictions
    cities = pd.DataFrame(predictions_test)
    cities.to_csv('D:/A_NAUSHAD/E/RESULTS/PRED/test_wavenet_' + df.iloc[:, a:a+1].columns[0] + '_S_pred.csv')

    df_pred=pd.DataFrame(X_test[:,:][win_length:])

    rev_trans=scaler.inverse_transform(df_pred)

    df_final=dataset[rev_trans.shape[0]*-1:]

    df_final[[df.iloc[:,a:a+1].columns[0]+"_pred"]]=predictions_test

    df_final


    df_final[[df.iloc[:,a:a+1].columns[0],df.iloc[:,a:a+1].columns[0]+"_pred"]].plot(figsize=(20,10))

    def evaluate_preds(y_true, y_pred):
        y_true = tf.cast(y_true, dtype=tf.float32)
        y_pred = tf.cast(y_pred, dtype=tf.float32)

        # Calculate various metrics
        mae = tf.keras.metrics.mean_absolute_error(y_true, y_pred)
        mse = tf.keras.metrics.mean_squared_error(y_true, y_pred)
        rmse = tf.sqrt(mse)
        mape = tf.keras.metrics.mean_absolute_percentage_error(y_true, y_pred)
        mslr = tf.keras.metrics.mean_squared_logarithmic_error(y_true, y_pred)

        return {"mae": mae.numpy(),
                "mse": mse.numpy(),
                "rmse": rmse.numpy(),
                "mape": mape.numpy(),
                "mslr": mslr.numpy(),
               }


    selected_features = [df.iloc[:,a:a+1].columns[0]]

    for feature in selected_features:
        print("-----{}-----".format(feature))
#         evala=evaluate_preds(y_true=df_final[feature].values,
#                              y_pred=df_final["{}_pred".format(feature)].values)
#         print(evala)
#         eval=pd.DataFrame.from_dict(evala,index=[0] )
#         eval.to_csv(df.iloc[:,a:a+1].columns[0]+"_evaluation.csv")
        eval_results = evaluate_preds(y_true=df_final[feature].values, y_pred=df_final["{}_pred".format(feature)].values)
        eval_df = pd.DataFrame.from_dict(eval_results, orient='index', columns=['value'])
        eval_df.to_csv('D:/A_NAUSHAD/E/RESULTS/EVAL\{}_wavenet_eval_S.csv'.format(df.iloc[:, a:a+1].columns[0]))
```
